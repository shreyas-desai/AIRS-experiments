# Temporal symetry
if df['source'] == df['target'] and df['target'] == df['source'] and df['relation'] == df ['relation'] and df['time']!=df['time']

```python
def rule_3(data, src, relation, target, date,text): # apply static symmetry
    grouped = data.groupby(by = [relation])[[src, target, date, text]].apply(lambda x:x).droplevel(1)
    grouped = grouped.reset_index()
    grouped.drop_duplicates(subset=['source','target'], inplace=True)
    merged = pd.merge(grouped, grouped, left_on=['source', 'target', "relation"], right_on=['target', 'source',"relation"])
    merged.drop(merged.filter(regex='_y$').columns, axis=1, inplace=True)
    merged.rename(columns = {"source_x":"source","target_x":"target"}, inplace=True)
    return merged
```
